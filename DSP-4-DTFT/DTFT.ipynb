```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, ifft, fftshift
import cmath
```

## Discrete Time Fourier Transform

```python
def DTFT(x,N):
    input_length = len(x)
    y = []
    w = np.linspace(-np.pi,np.pi,N)

    for k in range(N):  
        s = 0
        for n in range(input_length):
            s += x[n] * cmath.exp(-1j * w[k] * n)
        y.append(s)
    return y
```

```python
def plotDTFT(y, color):
    samples = 100001
    x = np.linspace(-np.pi,np.pi,samples)
    y_fft = DTFT(y, samples)
    y_fft_magnitude = np.absolute(y_fft)
    y_fft_phase = np.angle(y_fft)
    
    plot1 = plt.figure(1)
    plt.plot(x, y_fft_magnitude, color = color)
    plt.title('DTFT (Magnitude)')
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency (radians)')

    plot2 = plt.figure(2)
    plt.plot(x, y_fft_phase, color = color)
    plt.title('DTFT (Phase)')
    plt.ylabel('Phase')
    plt.xlabel('Frequency (radians)')
```

#### Sine generator

```python
def sine(f,phase, size):
    x = np.linspace(0,size,size).astype(int)
    y = np.sin(f*2*np.pi*x + phase)[:size + 1]
    return x, y
```

### I. Amplitude = 1, Phase = 0, Frequency = 0.1, Length = 100.

```python
[t, y] = sine(0.1, 0, 100)

plt.plot(t,y, color = "blue")
plt.xlim([-50, 150])
```

```python
plotDTFT(y, color = "blue")
```
