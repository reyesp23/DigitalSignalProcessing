```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile as reader
from scipy import signal
import sounddevice as sd
```

## Cross-correlation

#### Song

```python
#Read audio file
fs_song, y_song = reader.read('./Corcovado.wav')

#Take single channel
mono_song = y_song[:,0].astype(np.float64)

plot1 = plt.figure(1)
plt.plot(mono_song, color = "purple")
plt.title('Corcovado')

plt.show()
sd.play(y_song.astype('int16'))
```

#### Probe

```python
#Read audio file
fs_probe, y_probe = reader.read('./clip.wav')

#Take single channel
mono_probe = y_probe[:,0].astype(np.float64)

plot1 = plt.figure(1)
plt.plot(mono_probe, color = "orange")
plt.title('Corcovado clip')
plt.show()
sd.play(y_probe.astype('int16'))
```

### Cross-correlation

```python
result = signal.correlate(mono_song, mono_probe, 'valid')

plot1 = plt.figure(1)
plt.plot(result, color = "green")

plt.title('Cross-correlation')
plt.show()
```

```python
maxtime=(np.argmax(result))/fs_song
```

```python
import math
start_time_hours = math.floor(maxtime/60)
start_time_seconds = maxtime % 60
print("Start time {}:{}".format(start_time_hours, start_time_seconds))
```
