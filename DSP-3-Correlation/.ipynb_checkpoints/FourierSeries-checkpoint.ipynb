```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Fourier Series

$$ y( t) \ =\ \sum _{k=1}^{K}\frac{sin( 2\pi ( 2k-1) ft)}{2k-1} ,\ \rightarrow \ K=\infty  $$

```python
def harmonicsGenerator(k):
    f = 2
    T = np.linspace(0, 1, 1000)
    result = []
    for t in T:
        result.append(((np.sin((2*k - 1)*2*np.pi*f*t))/(2*k-1)))
    return result
```

```python
def sine(k):
    res = harmonicsGenerator(1)
    for i in range(2, k+1):
        res = res + np.array(harmonicsGenerator(i)) 
    return res
```

#### Plot multiple harmonics

```python
fig = plt.figure()
plt.plot(harmonicsGenerator(1))
plt.plot(harmonicsGenerator(3))
plt.plot(harmonicsGenerator(5))
plt.plot(harmonicsGenerator(10))
plt.title('Harmonics')
```

#### Plot fourier series for different k values

```python
k = 1
f = sine(k)
plt.plot(f)
plt.title('Fourier Series (k = 1)')
```

```python
f_1 = sine(1)
f_3 = sine(3)
f_9 = sine(9)
f_50 = sine(50)
f_100 = sine(100)
plt.plot(f_1, label='k = 1')
plt.plot(f_3, label='k = 3')
plt.plot(f_9, label='k = 9')
plt.plot(f_50, label='k = 50')
plt.plot(f_100, label='k = 100')
plt.legend()
plt.title('Fourier Series (combined)')
```
